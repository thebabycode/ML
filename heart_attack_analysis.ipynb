Description:

Age : Age of the patient

Sex : Sex of the patient

exng: exercise induced angina (1 = yes; 0 = no)

caa: number of major vessels (0-3)

cp : Chest Pain type chest pain type

  Value 1: typical angina
  Value 2: atypical angina
  Value 3: non-anginal pain
  Value 4: asymptomatic
trtbps : resting blood pressure (in mm Hg)

oldpeak : ST depression induced by exercise relative to rest

chol : cholestoral in mg/dl fetched via BMI sensor

fbs : (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)

rest_ecg : resting electrocardiographic results

  Value 0: normal
  Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
  Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
thalach : maximum heart rate achieved

##Implementation of Logistic Regression

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X = df[['restecg','slp','thalachh','cp','oldpeak']]
y = df['output']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
scaler = StandardScaler()

scaled_x_train = scaler.fit_transform(X_train)
scaled_x_test = scaler.transform(X_test)

scaled_x_train
```

```python
import sklearn.linear_model as sk
```

##Implemtation of KNN and Confusion Matrix

```python
from sklearn.metrics import classification_report , plot_confusion_matrix ,plot_precision_recall_curve ,plot_roc_curve
```

```python
from sklearn.neighbors import  KNeighborsClassifier
```

```python
from sklearn.metrics import accuracy_score
import sklearn.metrics as metrics
```

##KNN using EUCLEDEAN

```python
y1=[]
for i in range(2,6):
    knn1 = KNeighborsClassifier(n_neighbors=i,metric='euclidean')
    knn1.fit(scaled_x_train,y_train)
    pred1 = knn1.predict(scaled_x_test)
    print(f"For Knn-{i}: \n")
    print(classification_report(y_test,pred1))
    print("Accuracy Score for K",i,"is:", accuracy_score(y_test, pred1)*100, "%")
    y1.append(metrics.accuracy_score(y_test, pred1)*100)
    print('*************************************')

plt.scatter(range(2,6),y1)
plt.figure(figsize=(2, 3))
plt.show
```

##KNN using Minkowski

```python
y2=[]
for i in range(2,6):

    knn1 = KNeighborsClassifier(n_neighbors=i,metric='minkowski')
    knn1.fit(scaled_x_train,y_train)
    pred1 = knn1.predict(scaled_x_test)
    print(f"For Knn-{i}: \n")
    print(classification_report(y_test,pred1))
    print("Accuracy Score for K",i,"is:", accuracy_score(y_test, pred1)*100, "%")
    y2.append(metrics.accuracy_score(y_test, pred1)*100)
    print('*************************************')

plt.scatter(range(2,6),y2)
plt.figure(figsize=(2, 3))
plt.show
```

##KNN using Manhattan

```python
from sklearn import metrics
y3=[]
for i in range(2,6):
    knn1 = KNeighborsClassifier(n_neighbors=i,metric='manhattan')
    knn1.fit(scaled_x_train,y_train)
    pred1 = knn1.predict(scaled_x_test)
    print(f"For Knn-{i}: \n")
    print(classification_report(y_test,pred1))
    print("Accuracy Score for K",i,"is:", accuracy_score(y_test, pred1)*100, "%")
    y3.append(metrics.accuracy_score(y_test, pred1)*100)
    print('*************************************')
plt.scatter(range(2,6),y3)
plt.figure(figsize=(2, 3))
plt.show
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(scaled_x_train, y_train)
predicted=model.predict(scaled_x_test)
conf = confusion_matrix(y_test, predicted)
sns.set(rc = {'figure.figsize':(8,5)})
sns.heatmap(data=conf,annot=True)
print ("Confusion Matrix : \n", conf)
print ("The accuracy of Logistic Regression is : ", accuracy_score(y_test, predicted)*100, "%")
```

##LAB 4

##K-Means

```python
# x and y given as DataFrame columns
import plotly.express as px
fig = px.scatter(df, x='chol', y="trtbps", color='output', symbol="sex")
fig.show()
```

```python
import plotly.graph_objects as go
from sklearn.cluster import KMeans

wcss = []
clus = 6
for i in range(1, clus):
    kmeans = KMeans(n_clusters = i, init = "k-means++", max_iter = 500, n_init = 10, random_state = 123)
    kmeans.fit(X_train.values)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,clus),wcss)
plt.show()
    
fig = go.Figure(data = go.Scatter(x = [i for i in range(1, clus)], y = wcss))


fig.update_layout(title='WCSS vs. Cluster number',
                   xaxis_title='Clusters',
                   yaxis_title='WCSS')
fig.show()
```

```python
model=KMeans(n_clusters=6)
model.fit(df[['age','chol','trtbps']])
model.labels_
df['Cluster Number']=model.labels_
df
```

```python
fig = px.scatter(df, x='chol', y="trtbps", color=model.labels_, symbol="output")
fig.show()
```

```python
model=KMeans(n_clusters=2)
model.fit(df[['age','chol','trtbps']])
model.labels_
df['Cluster Number']=model.labels_
df
```

```python
fig = px.scatter(df, x='chol', y="trtbps", color=model.labels_, symbol="output")
fig.show()
```

Executing a method that returns some important key values of Linear Regression

```python
slope, intercept, r, p, std_err = stats.linregress(x1, y1)

def myfunc(x1):
  return slope * x1 + intercept
```

```python
plt.scatter(x1, y1)
```

```python
mymodel = list(map(myfunc, x1))
plt.scatter(x1, y1)
plt.plot(x1, mymodel)
plt.show()
```
