```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = data
```

```python
# correlation matrix heatmap visualization
sns.set(style="white")

fig, ax = plt.subplots(figsize=(15,11))

_ = sns.heatmap(df.corr(), annot=True, annot_kws={"size": 12}, square=True, 
                cmap='coolwarm' , vmin=-1, vmax=1, fmt='.2f')  
```

```python
X = df[['clock_speed','battery_power','n_cores','ram','int_memory']]
y = df['price_range']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
dt = DecisionTreeClassifier(max_depth = 3)
dt.fit(X_train, y_train)
```

```python
plt.figure(figsize=(27,14))
plot_tree(dt, feature_names = X.columns, class_names=["0", "1","2","3"], filled=True);
```
